```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('datasets/haberman.csv')
```

```python
data.shape
```

```python
data['survival'].value_counts()
```

```python
data.head()
```

```python
data = data.rename(columns={'30':'age','64':'opYears','1':'axilNodes','1.1':'survival'})
```

```python
data.plot(kind = 'scatter',x = 'age',y = 'survival')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.pairplot(data,size=3)
plt.show()
```

```python
data.plot(kind='scatter',x='age',y='axilNodes')
```

```python
plt.close()
sns.FacetGrid(data,hue='survival',size=5)\
    .map(sns.displot, 'axilNodes')\
        .add_legend()
plt.show()
```

```python
sns.FacetGrid(data,hue='survival',size=5)\
    .map(sns.displot, 'age')\
        .add_legend()
plt.show()
```

```python
sns.FacetGrid(data,hue='survival',size=5)\
    .map(sns.displot, 'opYears')\
        .add_legend()
plt.show()
```

```python
print(np.mean(data['age']))
print(np.std(data['age']))
```

```python
plt.close()
sns.set_style("whitegrid")
sns.FacetGrid(data, hue='survival',height=6).map(sns.distplot,'age').add_legend();
```
